```python
import random


def genGrid():
    grid = []
    for row in range(3):
        grid.append([' ']*3)
    return grid


def printGrid(grid):
    colNumbers = "  "
    for colNum in range(3):
        colNumbers += str(colNum + 1) + "   "
    print(colNumbers)
    for index,row in enumerate(grid):
        rowOutput = str(index + 1) + " "
        for cellNum,cell in enumerate(row):
            rowOutput += cell + " | "
        print(rowOutput)


def userInput(grid):
    while True:
        try:
            guess = input("Enter your next position <x,y>: ")
            guess = guess.split(',')
            guessX = int(guess[1])
            guessY = int(guess[0])
            if not (guessX > 0 and guessX <= 3) and (guessY > 0 and guessY <= 3):
                print("MAX co-ordinate value is " + str(3))
            elif grid[guessX - 1][guessY - 1] != " ":
                print("This position is occupied!")
            else:
                break
        except:
            print("Your Selection should be like: x,y")
            
    grid[guessX - 1][guessY - 1] = "X"
    return guessX, guessY

def winCheck(grid, player):
    return (
        (grid[0][0] == player and grid[1][0] == player and grid[2][0] == player) or 
        (grid[0][1] == player and grid[1][1] == player and grid[2][1] == player) or 
        (grid[0][2] == player and grid[1][2] == player and grid[2][2] == player) or 

        (grid[0][0] == player and grid[0][1] == player and grid[0][2] == player) or 
        (grid[1][0] == player and grid[1][1] == player and grid[1][2] == player) or 
        (grid[2][0] == player and grid[2][1] == player and grid[2][2] == player) or 

        (grid[0][0] == player and grid[1][1] == player and grid[2][2] == player) or 
        (grid[2][0] == player and grid[1][1] == player and grid[0][2] == player) 
    )


def copyGrid(grid):
    copy = []
    for index,row in enumerate(grid):
        copy.append([])
        for cell in row:
            copy[index].append(cell)
    return copy

def aiTurn(grid):
    
    for rowNum in range(0,3):
        for cellNum in range(0,3):
            copy = copyGrid(grid)
            if copy[rowNum][cellNum] == " ":
                copy[rowNum][cellNum] = "O"
                if winCheck(copy, "O"):
                    return rowNum, cellNum

   
   
    for rowNum in range(0,3):
        for cellNum in range(0,3):
            copy = copyGrid(grid)
            if copy[rowNum][cellNum] == " ":
                copy[rowNum][cellNum] = "X"
                if winCheck(copy, "X"):
                    return rowNum, cellNum

    
    possibleMoves = []
    if grid[0][0] == " ":
        possibleMoves.append([0,0])
    if grid[2][0] == " ":
        possibleMoves.append([2,0])
    if grid[0][2] == " ":
        possibleMoves.append([0,2])
    if grid[2][2] == " ":
        possibleMoves.append([2,2])

    if len(possibleMoves) > 0:
        move = random.choice(possibleMoves)
        return move[0], move[1]

    
    if grid[1][1] == " ":
        return 1,1

   
    possibleMoves = []
    if grid[1][0] == " ":
        possibleMoves.append([1,0])
    elif grid[2][1] == " ":
        possibleMoves.append([2,1])
    elif grid[1][2] == " ":
        possibleMoves.append([1,2])
    elif grid[0][1] == " ":
        possibleMoves.append([0,1])

    if len(possibleMoves) > 0:
        move = random.choice(possibleMoves)
        return move[0], move[1]

    
    return None

if __name__ == "__main__":
    grid = genGrid()
    printGrid(grid)

    count = 0 
    while not winCheck(grid, "X") and not winCheck(grid, "O"):  
        if count >= 9:  
            break
        count += 1
        print("\n")

        userInput(grid)
        printGrid(grid)


        if not winCheck(grid, "X"):    
            if count >= 9:  
                break
            count += 1
            print("\n")

            print("Computers turn...")
            aiRow, aiCol = aiTurn(grid)    
            grid[aiRow][aiCol] = "O"    
            printGrid(grid)     
        else:
            break       

    if winCheck(grid, "X"):
        print("You Win!!")
    elif winCheck(grid, "O"):
        print("Computer Wins!!")
    else:
        print("It's Over!!")
```
